## QuickStart `dlt pipline` with `pipeline_name`, `destination` and `dataset_name`

````python
import dlt

# a quick start data pipeline
pipeline = dlt.pipeline(
    pipeline_name="QuickStart",
    destination="duckdb",
    dataset_name="mydata"
)
````

### To load the data, you call the `run()` method and pass your data in my case its
```
data = [
    {"id": "1", "name": "bulbasaur", "size": {"weight": 6.9, "height": 0.7}},
    {"id": "4", "name": "charmander", "size": {"weight": 8.5, "height": 0.6}},
    {"id": "25", "name": "pikachu", "size": {"weight": 6, "height": 0.4}},
]
```
### in the data argument

````python
load_info = pipeline.run(data, table_name="Pokemon")
print(load_info)
````

# Explore the loaded data

## Duckdb Connection



````python
import duckdb
from google.colab import data_table
data_table.enable_dataframe_formatter()
# connection to duckdb database
conn = duckdb.connect(f"{pipeline.pipeline_name}.duckdb")
# Set serach path for the data_set
conn.sql(f"SET search_path = '{pipeline.dataset_name}'")

conn.sql("DESCRIBE").df()
````

````python
table = conn.sql("SELECT * FROM pokemon").df()
table
````

````python
with pipeline.sql_client() as client:
  with client.execute_query("SELECT * FROM pokemon") as cursor:
    table = cursor.df()

table
````

````python
dataset = pipeline.dataset(dataset_type="default")
dataset.pokemon.df()
````
